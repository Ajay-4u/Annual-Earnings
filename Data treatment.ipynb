# Data Treatment

Converting the object columns to int64

This would also remove any missing value from the data

```python
# data conversion for workclass column
private = adult[adult['workclass']=='Private']
private.loc[:, 'workclass'] = 1

seni = adult[adult['workclass']=='Self-emp-not-inc']
seni.loc[:, 'workclass'] = 2

sei = adult[adult['workclass']=='Self-emp-inc']
sei.loc[:, 'workclass'] = 3

fg = adult[adult['workclass']=='Federal-gov']
fg.loc[:, 'workclass'] = 4

lg = adult[adult['workclass']=='Local-gov']
lg.loc[:, 'workclass'] = 5

sg = adult[adult['workclass']=='State-gov']
sg.loc[:, 'workclass'] = 6

wp = adult[adult['workclass']=='Without-pay']
wp.loc[:, 'workclass'] = 7

nw = adult[adult['workclass']=='Never-worked']
nw.loc[:, 'workclass'] = 8

adult1= pd.concat([private, seni, sei, fg, lg, sg, wp, nw])
```

```python
# data conversion for education column
bach = adult1[adult1['education']=='Bachelors']
bach.loc[:, 'education'] = 1

sc = adult1[adult1['education']=='Some-college']
sc.loc[:, 'education'] = 2

e11 = adult1[adult1['education']=='11th']
e11.loc[:, 'education'] = 3

hs = adult1[adult1['education']=='HS-grad']
hs.loc[:, 'education'] = 4

ps = adult1[adult1['education']=='Prof-school']
ps.loc[:, 'education'] = 5

aa = adult1[adult1['education']=='Assoc-acdm']
aa.loc[:, 'education'] = 6

av = adult1[adult1['education']=='Assoc-voc']
av.loc[:, 'education'] = 7

n9 = adult1[adult1['education']=='9th']
n9.loc[:, 'education'] = 8

s7e8 = adult1[adult1['education']=='7th-8th']
s7e8.loc[:, 'education'] = 9

t12 = adult1[adult1['education']=='12th']
t12.loc[:, 'education'] = 10

mast = adult1[adult1['education']=='Masters']
mast.loc[:, 'education'] = 11

f1f4 = adult1[adult1['education']=='1st-4th']
f1f4.loc[:, 'education'] = 12

t10 = adult1[adult1['education']=='10th']
t10.loc[:, 'education'] = 13

doc = adult1[adult1['education']=='Doctorate']
doc.loc[:, 'education'] = 14

f5s6 = adult1[adult1['education']=='5th-6th']
f5s6.loc[:, 'education'] = 15

pschool = adult1[adult1['education']=='Preschool'] 
pschool.loc[:, 'education'] = 16

adult2= pd.concat([bach, sc, e11, hs, ps, aa, av, n9, s7e8, t12, mast, f1f4, t10, doc, f5s6, pschool])
```

```python
adult2.nunique()
```

```python
# data conversion for marital status column
mcs = adult2[adult2['marital_status']=='Married-civ-spouse']
mcs.loc[:, 'marital_status'] = 1

div = adult2[adult2['marital_status']=='Divorced']
div.loc[:, 'marital_status'] = 2

nm = adult2[adult2['marital_status']=='Never-married']
nm.loc[:, 'marital_status'] = 3

sep = adult2[adult2['marital_status']=='Separated']
sep.loc[:, 'marital_status'] = 4

wid = adult2[adult2['marital_status']=='Widowed']
wid.loc[:, 'marital_status'] = 5

msa = adult2[adult2['marital_status']=='Married-spouse-absent']
msa.loc[:, 'marital_status'] = 6

mas = adult2[adult2['marital_status']=='Married-AF-spouse']
mas.loc[:, 'marital_status'] = 7

adult3= pd.concat([mcs, div, nm, sep, wid, msa, mas])
```

```python
# data conversion for occupation column
tech = adult3[adult3['occupation']=='Tech-support']
tech.loc[:, 'occupation'] = 1

cr = adult3[adult3['occupation']=='Craft-repair']
cr.loc[:, 'occupation'] = 2

os = adult3[adult3['occupation']=='Other-service']
os.loc[:, 'occupation'] = 3

sales = adult3[adult3['occupation']=='Sales']
sales.loc[:, 'occupation'] = 4

em = adult3[adult3['occupation']=='Exec-managerial']
em.loc[:, 'occupation'] = 5

profspe = adult3[adult3['occupation']=='Prof-specialty']
profspe.loc[:, 'occupation'] = 6

hc = adult3[adult3['occupation']=='Handlers-cleaners']
hc.loc[:, 'occupation'] = 7

moi = adult3[adult3['occupation']=='Machine-op-inspct']
moi.loc[:, 'occupation'] = 8

ac = adult3[adult3['occupation']=='Adm-clerical']
ac.loc[:, 'occupation'] = 9

ff = adult3[adult3['occupation']=='Farming-fishing']
ff.loc[:, 'occupation'] = 10

tm = adult3[adult3['occupation']=='Transport-moving']
tm.loc[:, 'occupation'] = 11

phs = adult3[adult3['occupation']=='Priv-house-serv']
phs.loc[:, 'occupation'] = 12

pros = adult3[adult3['occupation']=='Protective-serv']
pros.loc[:, 'occupation'] = 13

af = adult3[adult3['occupation']=='Armed-Forces']
af.loc[:, 'occupation'] = 14

adult4= pd.concat([tech,cr,os,sales,em,profspe,hc,moi,ac,ff,tm,phs,pros,af])
```

```python
# data conversion for relationship column
wife = adult4[adult4['relationship']=='Wife']
wife.loc[:, 'relationship'] = 1

ochild = adult4[adult4['relationship']=='Own-child']
ochild.loc[:, 'relationship'] = 2

husband = adult4[adult4['relationship']=='Husband']
husband.loc[:, 'relationship'] = 3

nif = adult4[adult4['relationship']=='Not-in-family']
nif.loc[:, 'relationship'] = 4

orela = adult4[adult4['relationship']=='Other-relative']
orela.loc[:, 'relationship'] = 5

unmar = adult4[adult4['relationship']=='Unmarried']
unmar.loc[:, 'relationship'] = 6

adult5= pd.concat([wife, ochild, husband, nif, orela, unmar])
```

```python
# data conversion for race column
white = adult5[adult5['race']=='White']
white.loc[:, 'race'] = 1

api = adult5[adult5['race']=='Asian-Pac-Islander']
api.loc[:, 'race'] = 2

aie = adult5[adult5['race']=='Amer-Indian-Eskimo']
aie.loc[:, 'race'] = 3

other = adult5[adult5['race']=='Other, ']
other.loc[:, 'race'] = 4

black = adult5[adult5['race']=='Black']
black.loc[:, 'race'] = 5

adult6= pd.concat([white, api, aie, other, black])
```

```python
# data conversion for sex column
female = adult6[adult6['sex']=='Female']
female.loc[:, 'sex'] = 0

male = adult6[adult6['sex']=='Male']
male.loc[:, 'sex'] = 1

adult7= pd.concat([female, male])
```

```python
# data conversion for earnings column
no = adult7[adult7['earnings']=='<=50K']
no.loc[:, 'earnings'] = 0

yes = adult7[adult7['earnings']=='>50K']
yes.loc[:, 'earnings'] = 1

adult8= pd.concat([no, yes])
```
